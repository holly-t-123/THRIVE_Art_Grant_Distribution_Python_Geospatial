```python
thrive = pd.read_csv('../data/Metro_Arts_THRIVE_Projects_and_Locations.csv')
print(thrive.shape)
thrive
```

```python
thrive.info()
```

```python
thrive['geometry'] = thrive.apply(lambda x: Point(x.Longitude,x.Latitude), 
                                  axis=1)

thrive.head()
```

```python
council_districts.crs
```

```python
art_geo = gpd.GeoDataFrame(thrive, crs=council_districts.crs, geometry=thrive.geometry)
```

```python
type(art_geo)
```

```python
council_districts = council_districts[['council_district','last_name','geometry']]
```

```python
art_by_district = gpd.sjoin(art_geo, council_districts, op='within')
```

```python
art_by_district.council_district.count()
```

```python
#less than half of the 35 council districts have a THRIVE art installation
art_by_district['council_district'].value_counts()
```

```python
proj_in_dis19 = art_by_district[art_by_district.council_district == '19']
proj_in_dis19.shape
```

```python
polygon19 = council_districts[council_districts.council_district == '19']
polygon19.head()
```

```python
ax = polygon19.plot(figsize=(8,10), color='lightgreen')
proj_in_dis19.plot(ax=ax, column = 'Organization/Artist Name')
```

```python
center = polygon19.geometry.centroid
print(center)
```

```python
#indexing at 18 to just get the coordinates
center = polygon19.geometry.centroid[18]
print(center)
```

```python
area_center = [center.y, center.x]
print(area_center)
```

```python
map19 = folium.Map(location = area_center, zoom_start = 12)
map19
```

```python
i = 0
for ind, row in proj_in_dis19.iterrows():
    if i < 11:
        print('index is', ind)
        print('values are:\n')
        print(row)
        print('-----------------------')
        i += 1
```

```python
map19 = folium.Map(location = area_center, zoom_start = 12)

folium.GeoJson(polygon19).add_to(map19)

for ind, row in proj_in_dis19.iterrows():
    loc = [row['Latitude'], row['Longitude']]
    pop = row['Organization/Artist Name']
    icon = folium.Icon(color='purple', icon = 'paint-brush', prefix='fa')
    
    marker = folium.Marker(location = loc, popup = pop, icon = icon)
    marker.add_to(map19)
    
map19.save('../maps/map19.html')
map19
```

```python
cluster_map19 = folium.Map(location = area_center, zoom_start = 12)

marker_cluster = MarkerCluster().add_to(cluster_map19)

folium.GeoJson(polygon19).add_to(cluster_map19)

for ind, row in proj_in_dis19.iterrows():
    loc = [row['Latitude'], row['Longitude']]
    pop = row['Organization/Artist Name']
    icon = folium.Icon(color='purple', icon = 'paint-brush', prefix = 'fa')
    
    marker = folium.Marker(location = loc, popup = pop, icon = icon)
    marker.add_to(marker_cluster)
    
cluster_map19.save('../maps/cluster19.html')
cluster_map19
```

```python
map19 = folium.Map(location = area_center, zoom_start = 12)
folium.GeoJson(council_districts).add_to(map19)
marker_cluster = MarkerCluster().add_to(map19)

for ind, row in art_by_district.iterrows():
    loc = [row['Latitude'],row['Longitude']]
    pop = 'Artist/Organization: ' + row['Organization/Artist Name'] + '\n\nCouncil District: ' + row['council_district'] + ' (' + row['last_name'] + ')'
    icon = folium.Icon(color='purple', icon = 'paint-brush', prefix='fa')
    
    marker = folium.Marker(location = loc, popup = pop, icon = icon)
    
    marker.add_to(marker_cluster)
    
map19.save('../maps/cluster19.html')
map19
```
